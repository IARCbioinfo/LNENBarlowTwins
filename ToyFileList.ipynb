```python
import os
import random
```

```python
root = "/home/mathiane/LNEN_files_imgs/TumorSegmentationDataSets/TumorNonTumor_Ki67_dataset"
```

```python
flist = []
for tne in os.listdir(root):
    for tumor_seg in os.listdir(os.path.join(root, tne)):
        for tile in os.listdir(os.path.join(root,tne, tumor_seg)):
            flist.append(os.path.join(root,tne, tumor_seg, tile))
```

```python
flist
```

```python
len(flist)
```

```python
random.shuffle(flist)
```

```python
with open("Datasets/ToyTrainingSetKi67Tumor.txt", "w") as f:
    for ele in flist[:5000]:
        f.write(ele+"\n")
with open("Datasets/ToyTestSetKi67Tumor.txt", "w") as f:
    for ele in flist[5000:7000]:
        f.write(ele+"\n")
```
